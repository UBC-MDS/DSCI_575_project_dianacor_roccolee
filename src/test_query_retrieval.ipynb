```python
from utils import *
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer
index = faiss.read_index("../data/processed/semantic_search_index.faiss")
```

```python
test_queries = pd.read_csv("../data/processed/test_queries.csv")
model = SentenceTransformer("all-MiniLM-L6-v2")
docs = pd.read_parquet("../data/processed/product_documents.parquet")
```

```python
master_retrieval = []

for query in test_queries['queries']:
    result = semantic_search(docs, model, index, query)
    for r in result:
        master_retrieval.append({
            'query': query,
            'semantic_search_results': r['product_title']
        })

master_retrieval = pd.DataFrame(master_retrieval)
master_retrieval
```
