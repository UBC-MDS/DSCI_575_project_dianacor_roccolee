```python
from utils import *
import pickle
import pandas as pd
from rank_bm25 import BM25Okapi


df = pd.read_parquet("../data/processed/product_documents.parquet", columns=["document", "product_title"])
output_dir = "data/processed/"

tokenized_corpus = [tokenize(text) for text in df["document"]]
doc_names = df["product_title"].tolist()

print("Building BM25 index")
bm25 = BM25Okapi(tokenized_corpus)

print("Saving BM25 index and tokenized corpus")
with open(f"{output_dir}bm25_index.pkl", "wb") as file:
    pickle.dump({"bm25": bm25, "doc_names": doc_names}, file)

with open(f"{output_dir}tokenized_corpus.pkl", "wb") as file:
    pickle.dump(tokenized_corpus, file)

print(f"Saved bm25_index.pkl and tokenized_corpus.pkl in {output_dir}")
```
