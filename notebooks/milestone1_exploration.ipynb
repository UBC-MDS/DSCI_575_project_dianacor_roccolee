## Selection and justification of fields for retrieval

For the metadata table, we decided to select main_category, title, features, description and details as the fields we would include for retrieval. We chose these columns because each one adds a unique identifier to many of the devices in our category. The labels in main_category are often used in user queries if they aren't sure what they are looking for, or they are searching by function instead of by product. The title field narrows the search down a bit in case the user wants to search for a specific line of products, such as iPad or Security Camera with Flood Light. The remaining columns (features, description and details) all add more granular data to the document for each product if the user query includes specifics down to the specifications or a particular model of the device.

For the reviews table, we only included title and helpful_vote as required columns. We wanted to avoid adding the entire review because that would make the document too large to compute without adding too much valuable information. To avoid adding the entire review, we believe the title field will provide a good summary of the review's contents. The helpful_vote field was also included because we may have future plans to expand our retrieval system to consider the number of helpful votes of a recommended product, with the assumption that reviews with more helpful votes should have more weight in how good a product is. However, this feature may be out of scope for this milestone.  

## Description of Test Preprocessing Decisions

To start, the usual text preprocessing will be done, which will include removing stopwords, punctuation and whitespace before tokenizing the documents to build the corpus. For more specific columns, we have cleaned the input to maximize the amount of useful information and reduce noise. For example, we are only taking the first 100 characters of the title column in the metadata table because otherwise there would be a lot of noise in some of the product titles. We will also perform preprocessing on the features, description and details columns in the metadta table to flatten them into 1 string.
